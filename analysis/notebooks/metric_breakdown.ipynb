# Metric Breakdown Notebook
This notebook recomputes the aggregated metrics reported in Chapters~4 and 6 starting from `analysis_model_summary_table.csv`.

```python
import pandas as pd
from pathlib import Path
summary_path = Path('analysis_model_summary_table.csv')
df = pd.read_csv(summary_path)
display(df.head())
type_summary = df.groupby('Type')[['Composite','CLIPScore','BLEU1','METEOR','ROUGEL']].mean().sort_values('Composite', ascending=False)
type_summary.round(3)
```

The grouped dataframe above is exported to `analysis_type_averages.json` so the LaTeX tables can consume a consistent, versioned source.

```python
type_summary.reset_index().to_json('analysis_type_averages.json', orient='records', indent=2)
```
